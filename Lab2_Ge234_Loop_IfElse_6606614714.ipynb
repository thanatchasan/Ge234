## ðŸ”¹ à¹à¸šà¸šà¸à¸¶à¸à¸«à¸±à¸” 1: à¹€à¸‚à¸µà¸¢à¸™ for loop à¹€à¸žà¸·à¹ˆà¸­à¹à¸ªà¸”à¸‡à¸Šà¸·à¹ˆà¸­à¸ˆà¸±à¸‡à¸«à¸§à¸±à¸”à¹à¸¥à¸°à¸žà¸´à¸à¸±à¸”à¸ˆà¸²à¸à¸¥à¸´à¸ªà¸•à¹Œà¸‚à¸­à¸‡à¸„à¸¸à¸“à¹€à¸­à¸‡


```python
# à¸žà¸´à¸à¸±à¸”à¸‚à¸­à¸‡à¸ªà¸–à¸²à¸™à¸—à¸µà¹ˆà¸ªà¸³à¸„à¸±à¸à¹ƒà¸™à¸›à¸£à¸°à¹€à¸—à¸¨à¹„à¸—à¸¢
locations = {
    "Nan": (18.7756, 100.7730),
    'khon kaen': (16.4419, 102.8350),
    "Ayutthaya": (14.3532, 100.5689),
    "krabi": (8.0863, 98.9063)
}

# à¸§à¸™à¸¥à¸¹à¸›à¹à¸ªà¸”à¸‡à¸„à¹ˆà¸²à¸žà¸´à¸à¸±à¸”
for city, coord in locations.items():
    print(f"{city}: Latitude {coord[0]}, Longitude {coord[1]}")
```


## ðŸ”¹ à¹à¸šà¸šà¸à¸¶à¸à¸«à¸±à¸” 2: à¹ƒà¸Šà¹‰ if-else à¹€à¸žà¸·à¹ˆà¸­à¹à¸¢à¸à¹à¸¢à¸°à¸žà¸·à¹‰à¸™à¸—à¸µà¹ˆà¸—à¸µà¹ˆà¹€à¸›à¹‡à¸™à¸ à¸¹à¹€à¸‚à¸²à¹à¸¥à¸°à¸žà¸·à¹‰à¸™à¸—à¸µà¹ˆà¸£à¸²à¸šà¸ˆà¸²à¸à¸¥à¸´à¸ªà¸•à¹Œà¸‚à¸­à¸‡à¸žà¸´à¸à¸±à¸”


```python
for city, coord in locations.items():
    # à¸ªà¸¡à¸¡à¸•à¸´à¸à¸²à¸™: à¸¥à¸°à¸•à¸´à¸ˆà¸¹à¸”à¸—à¸µà¹ˆà¸ªà¸¹à¸‡à¸à¸§à¹ˆà¸² 17 à¸­à¸‡à¸¨à¸² à¸­à¸²à¸ˆà¹€à¸›à¹‡à¸™à¸žà¸·à¹‰à¸™à¸—à¸µà¹ˆà¸ à¸¹à¹€à¸‚à¸² (à¹‚à¸”à¸¢à¸—à¸±à¹ˆà¸§à¹„à¸›à¸ à¸²à¸„à¹€à¸«à¸™à¸·à¸­à¸‚à¸­à¸‡à¹„à¸—à¸¢)
    # à¸¥à¸°à¸•à¸´à¸ˆà¸¹à¸”à¸—à¸µà¹ˆà¸­à¸¢à¸¹à¹ˆà¸£à¸°à¸«à¸§à¹ˆà¸²à¸‡ 15.0 - 17.0 à¸­à¸‡à¸¨à¸² à¸­à¸²à¸ˆà¹€à¸›à¹‡à¸™à¸žà¸·à¹‰à¸™à¸—à¸µà¹ˆà¸£à¸²à¸šà¸ªà¸¹à¸‡ (à¹€à¸Šà¹ˆà¸™ à¸ à¸²à¸„à¸•à¸°à¸§à¸±à¸™à¸­à¸­à¸à¹€à¸‰à¸µà¸¢à¸‡à¹€à¸«à¸™à¸·à¸­)
    # à¹à¸¥à¸°à¸¥à¸°à¸•à¸´à¸ˆà¸¹à¸”à¸—à¸µà¹ˆà¸•à¹ˆà¸³à¸à¸§à¹ˆà¸²à¸«à¸£à¸·à¸­à¹€à¸—à¹ˆà¸²à¸à¸±à¸š 15.0 à¸­à¸‡à¸¨à¸² à¸­à¸²à¸ˆà¹€à¸›à¹‡à¸™à¸žà¸·à¹‰à¸™à¸—à¸µà¹ˆà¸£à¸²à¸šà¸«à¸£à¸·à¸­à¸Šà¸²à¸¢à¸à¸±à¹ˆà¸‡
    if coord[0] > 17.0:
        print(f"{city} : à¹€à¸›à¹‡à¸™à¸žà¸·à¹‰à¸™à¸—à¸µà¹ˆà¸ à¸¹à¹€à¸‚à¸²")
    elif coord[0] > 15.0 and coord[0] <= 17.0:
        print(f"{city} : à¹€à¸›à¹‡à¸™à¸—à¸µà¹ˆà¸£à¸²à¸šà¸ªà¸¹à¸‡")
    else:
        print(f"{city} : à¹€à¸›à¹‡à¸™à¸žà¸·à¹‰à¸™à¸—à¸µà¹ˆà¸£à¸²à¸šà¸«à¸£à¸·à¸­à¸Šà¸²à¸¢à¸à¸±à¹ˆà¸‡")
```


## ðŸ”¹ à¹à¸šà¸šà¸à¸¶à¸à¸«à¸±à¸” 3: à¹ƒà¸Šà¹‰ while loop à¹€à¸žà¸·à¹ˆà¸­à¸§à¸™à¸¥à¸¹à¸›à¹à¸¥à¸°à¹à¸ªà¸”à¸‡à¸„à¹ˆà¸²à¸¥à¸°à¸•à¸´à¸ˆà¸¹à¸”à¸ˆà¸²à¸à¸¥à¸´à¸ªà¸•à¹Œà¸ˆà¸™à¸à¸§à¹ˆà¸²à¸ˆà¸°à¸žà¸šà¸„à¹ˆà¸²à¸•à¹ˆà¸³à¸à¸§à¹ˆà¸² 10Â°


```python
points = [(18.7756, 100.7730), (16.4419, 102.8350), (14.3532, 100.5689), (8.0863, 98.9063)]
count = 0

while count < len(points) and points[count][0] >= 10:
    latitude = points[count][0]
    print(f"Latitude: {latitude}")
    count += 1

if count < len(points):
    print(f"à¸«à¸¢à¸¸à¸”à¸à¸²à¸£à¸§à¸™à¸¥à¸¹à¸›à¹€à¸™à¸·à¹ˆà¸­à¸‡à¸ˆà¸²à¸à¸žà¸šà¸¥à¸°à¸•à¸´à¸ˆà¸¹à¸”à¸•à¹ˆà¸³à¸à¸§à¹ˆà¸² 10 à¸­à¸‡à¸¨à¸²: Latitude {points[count][0]}")
```


## ðŸ”¹ à¹à¸šà¸šà¸à¸¶à¸à¸«à¸±à¸” 4: à¹ƒà¸Šà¹‰ for loop à¸§à¸´à¹€à¸„à¸£à¸²à¸°à¸«à¹Œà¸„à¹ˆà¸² NDVI à¹à¸¥à¸°à¸«à¸²à¸„à¹ˆà¸²à¸žà¸·à¹‰à¸™à¸—à¸µà¹ˆà¸—à¸µà¹ˆà¸¡à¸µà¸ªà¸ à¸²à¸žà¹à¸«à¹‰à¸‡à¹à¸¥à¹‰à¸‡ (NDVI < 0.2)


```python

import numpy as np

# à¸ˆà¸³à¸¥à¸­à¸‡à¸‚à¹‰à¸­à¸¡à¸¹à¸¥ NDVI (Normalized Difference Vegetation Index)
ndvi = np.array([
    [0.25, 0.35, 0.60],
    [0.05, 0.25, 0.75],
    [0.50, 0.85, 0.10]
])

# à¸§à¸™à¸¥à¸¹à¸›à¸•à¸£à¸§à¸ˆà¸ªà¸­à¸šà¸„à¹ˆà¸²à¸žà¸·à¸Šà¸žà¸£à¸£à¸“à¹€à¸žà¸·à¹ˆà¸­à¸«à¸²à¸žà¸·à¹‰à¸™à¸—à¸µà¹ˆà¹à¸«à¹‰à¸‡à¹à¸¥à¹‰à¸‡
for row in range(ndvi.shape[0]):
    for col in range(ndvi.shape[1]):
        if ndvi[row, col] < 0.2:
            print(f"à¸žà¸´à¸à¸±à¸” ({row}, {col}) à¸¡à¸µà¸ªà¸ à¸²à¸žà¹à¸«à¹‰à¸‡à¹à¸¥à¹‰à¸‡ (NDVI={ndvi[row, col]})")
```
